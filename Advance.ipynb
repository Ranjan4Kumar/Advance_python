```python
#write a program of inheritance



class org:
  def __init__(self, name, exp,dep):
    self.name = name
    self.exp = exp
    self.dep = dep

    print('(initialize emp {} :)'.format(self.name))


  def dept(self):
      print('Name of an employee {}, exp {} , dept {}'.format(self.name, self.exp, self.dep))


class empy(org):
  def __init__(self, name, exp, dep, age):
    org.__init__(self, name, exp, dep)
    self.age = age

  def dept(self):
    org.dept(self)
    print('Nmae is {}, exp is {:d}, dept is {}, age is {:d}'.format(self.name,self.exp,self.dep,self.age))


E1 = empy('rk', 12, 'devp', 25)

E1.dept()
```

```python

''' create a class an organisaton & add an employee to it 
making child class as dept and employee'''

#Class & Object

class Org:
  emp = 0

  def __init__(self, name , id):
    self.name = name
    self.id = id

    print('(Initilaize name of employee {})'.format(self.name))
    Org.emp += 1

  def delete(self):
        print('{} Left organization'.format(self.name))

        Org.emp -= 1

        if Org.emp == 0:
          print('{} was the last one'.format(self.name))
        else:

          print('There is still {:d} employees are working'.format(Org.emp))

  def say_hi(self):

    print('{} Joined today'.format(self.name))

  @classmethod
  def total_nos(cls):
    print('Total nos {:d}'.format(cls.emp))

E_1 = Org('ranjan', 184)
E_2 = Org('mitesh',178)
E_3 = Org('chandan', 185)
E_N = [E_1, E_2, E_3]
for i in E_N:
   i.total_nos()
E_1.delete()

```

```python
# Abstraction Method

''' in abstraction method we are calling module abc 
which we need to import from our library'''

from abc import ABC, abstractmethod

class Stand(ABC):
  def noofstudents(self):
    pass

class boys(Stand):
  def noofstudents(self):
    print('there are 40 boys')

class girls(Stand):
  def noofstudents(self):
    print('there are 50 girls')

students1 = boys()
students1.noofstudents()
students2 = girls()
students2.noofstudents()

```

```python
#interface
''' formal interface that is used to compare two classes in terms of 
true or false'''

from abc import ABC, abstractmethod
class kite(ABC):
  @abstractmethod
  def fly(self):
    pass

class rombus_shape(kite):
  def fly(self):
    print('It has four sides')

k = rombus_shape()
isinstance(k, kite)
```

```python
# creating another class

class bird(ABC):
  @abstractmethod
  def fly(self):
    pass

class sparrow(bird):
  def fly(self):
    print('flying')

v = sparrow()
isinstance(v, bird)
```

```python
isinstance(v, kite)
isinstance(k, bird)
```

```python
# Method - 2
# nested list hackerrank solution
# we need to create 2 lists 1 for marks and scores and 2nd one is specially for scores
# we ask user to give total how many entries he wants to give

n =int(input())
marksheet = []
scores = []
for i in range(n):
  name = input()
  score = float(input())
  marksheet.append([name,score])
  scores.append(score)
scores = sorted(set(scores))
m = scores[1]
name = []

for j in marksheet:
  if m == j[1]:
    name.append(j[0])
name.sort()

for k in name:
  print(k)
```
